```python
import pandas as pd
import pandas_datareader.data as web
from datetime import datetime

# Set the date range
start = datetime(2008, 1, 8)
end = datetime(2016, 7, 1)

# Retrieve daily data
interest_rate = web.DataReader("DFF", "fred", start, end)      # daily
exchange_rate = web.DataReader("DEXUSEU", "fred", start, end)  # daily (EUR/USD)

# Retrieve quarterly and monthly data
gdp_growth = web.DataReader("GDP", "fred", start, end)         # GDP quarterly
unemployment = web.DataReader("UNRATE", "fred", start, end)    # monthly
cpi = web.DataReader("CPIAUCSL", "fred", start, end)           # monthly
ppi = web.DataReader("PPIACO", "fred", start, end)             # monthly

# Forward-fill the monthly and quarterly data to create daily data
gdp_growth_daily = gdp_growth.resample('D').ffill()  # Resample to daily and forward-fill
unemployment_daily = unemployment.resample('D').ffill()
cpi_daily = cpi.resample('D').ffill()
ppi_daily = ppi.resample('D').ffill()

# Merge all data into a single DataFrame
merged_data = pd.concat([gdp_growth_daily, unemployment_daily, cpi_daily, ppi_daily, interest_rate, exchange_rate], axis=1)

# Rename columns for clarity
merged_data.columns = ['GDP', 'Unemployment', 'CPI', 'PPI', 'Interest Rate', 'Exchange Rate']

# Display the merged data
print(merged_data.head())


```
